```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
fdata['genres'] = fdata['genres'].apply(convert)
```

```python
fdata.iloc[0].genres
```

```python
fdata['keywords'] = fdata['keywords'].apply(convert)
```

```python
fdata.iloc[0].keywords
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
fdata['cast'] = fdata['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['department'] == "Directing":
            L.append(i['name'])
            break
    return L
```

```python
fdata['crew'] = fdata['crew'].apply(fetch_director)
```

```python
fdata.iloc[0].overview
```

```python
fdata['overview'] = fdata['overview'].apply(lambda x:x.split())
```

```python
fdata['genres']
```

```python
fdata['genres'] = fdata['genres'].apply(lambda x:[i.replace(" ","") for i in x])
fdata['keywords'] = fdata['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
fdata['cast'] = fdata['cast'].apply(lambda x:[i.replace(" ","") for i in x])
fdata['crew'] = fdata['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# fdata.head(1)
fdata['genres']
```

```python
fdata['tags'] = fdata['overview'] + fdata['genres'] + fdata['keywords'] + fdata['cast'] + fdata['crew'] 
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```
